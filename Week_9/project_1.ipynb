```python
import tkinter as tk
from random import choice

class EmployeeManagementSystem:
    employees = [
        "Mary Evans", "Eyo Ishan", "Durojaiye Dare", "Adams Ali", "Andrew Ugwu", 
        "Stella Mankinde", "Jane Akibo", "Ago James", "Michell Taiwo", "Abraham Jones", 
        "Nicole Anide", "Kosi Korso", "Adele Martins", "Emmanuel Ojo", "Ajayi Fatima"
    ]

    tasks = [
        "Loading", "Transporting", "Reviewing Orders", "Customer Service", "Delivering Items"
    ]

    def __init__(self, name):
        self.name = name

    def check_employee(self):
        return self.name in self.employees

    def take_attendance(self):
        if self.check_employee():
            return f"{self.name} is present for work today."
        else:
            return "Access Denied!"

    def assign_task(self):
        if self.check_employee():
            task = choice(self.tasks)
            return f"{self.name} has been assigned {task} for the day."
        else:
            return "Cannot assign task. Employee not found."

def main():
    root = tk.Tk()
    root.geometry('500x300')
    root.title("Employee Management System")

    frame = tk.Frame(root)
    frame.pack()

    tk.Label(frame, text="Enter your name:").pack()
    name_entry = tk.Entry(frame)
    name_entry.pack()

    def submit_name():
        name = name_entry.get()
        employee = EmployeeManagementSystem(name)
        attendance_result = employee.take_attendance()
        task_result = employee.assign_task()
        result_label.config(text=f"{attendance_result}\n{task_result}\n")

    tk.Button(frame, text="Submit", command=submit_name).pack()

    result_label = tk.Label(root)
    result_label.pack()

    root.mainloop()

if __name__ == "__main__":
    main()
```
